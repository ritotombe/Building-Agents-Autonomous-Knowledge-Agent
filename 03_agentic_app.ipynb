````python
from IPython.display import Image, display
from langchain_core.runnables.graph import MermaidDrawMethod

# Try different methods to generate the diagram
try:
    # Method 1: Try API method (most reliable in Jupyter)
    display(
        Image(
            orchestrator.get_graph().draw_mermaid_png(
                draw_method=MermaidDrawMethod.API
            )
        )
    )
except Exception as e:
    print(f"API method failed: {e}")
    try:
        # Method 2: Try without specifying method (uses default)
        display(
            Image(
                orchestrator.get_graph().draw_mermaid_png()
            )
        )
    except Exception as e2:
        print(f"Default method failed: {e2}")
        # Method 3: Fallback to Mermaid text
        print("Falling back to Mermaid text:")
        print("```mermaid")
        print(orchestrator.get_graph().draw_mermaid())
        print("```")
````
